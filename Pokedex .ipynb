```python
from playsound import playsound
from IPython.display import Image
import requests
from bs4 import BeautifulSoup
```

# Pokemon Classes and other lists

```python
mons=[]
st = ["Hit Points","Attack","Defense","Special","Speed"]
Types =["Normal","Fighting","Flying","Poison","Ground","Rock","Bug","Ghost","Fire","Water","Grass","Electric","Psychic","Ice","Dragon"]
class pokemon:
    def __init__(self,name,image,cry,types,moves,classi,b, cr,stats,h,w,n): 
        self.name= name
        self.img= image
        self.cry=cry
        self.type=types
        self.moves=list(dict.fromkeys(moves))
        self.classification=classi.lower()
        self.bio=b
        self.catch=cr
        self.stats= stats
        self.height=h
        self.weight=w
        self.num=n
```

# Implementing pokemons

```python
a = 1
while a <=151:
    web ='https://www.serebii.net/pokedex/'
    ntop = True
    t = 10
    b = 16
    sta=[]
    ty=[]
    mov=[]
    nm ="{}".format(a) 
    img =r"C:\\Users\\(ENTER NAME OF YOUR PC)\\Downloads\\pkmn sprites\\animated\\"+ nm+".gif" ##Rename to to work for your pc
    if len(nm) < 3:
        if len(nm) ==1:
            nm = '00'+nm
        elif len(nm) ==2:
            nm = '0'+nm
    r = requests.get(web+nm+'.shtml')
    soup = BeautifulSoup(r.text,'html.parser')
    results = soup.find_all('td',attrs={"class":"fooinfo"})
    nme = results[1].contents[0]
    nme1 = nme
    cl = results[4].contents[0]
    cr=int(results[7].contents[0])
    if a == 29:
        nme ='Nidoran-f'
        nme1 = 'Nidoran♀'
    elif a == 32:
        nme ='Nidoran-m'
        nme1='Nidoran♂'
    if a == 122:
        request_db = requests.get('https://pokemondb.net/pokedex/mr-mime')
        request_bulb = requests.get('https://bulbapedia.bulbagarden.net/wiki/Mr._Mime_(Pokemon)')
    else:
        request_db = requests.get('https://pokemondb.net/pokedex/'+nme.lower())
        request_bulb = requests.get('https://bulbapedia.bulbagarden.net/wiki/'+nme1+'_(Pokemon)')
    mfil=nm+"-"+nme.lower()+'.wav'
    mfil=r"C:\\Users\\(ENTER NAME OF YOUR PC)\\Downloads\\pokemon cries\\"+mfil ##Rename to to work for your pc
    #print(nme)
    si = BeautifulSoup(request_bulb.text,'html.parser')
    bio_info = si.find_all('td',attrs={"class":"roundy"})
    s = BeautifulSoup(request_db.text,'html.parser')
    type_info = s.find_all('td')
    biol = bio_info[41].text
    #print(biol)
    for str in Types:
        if str in type_info[1].text:
            ty.append(str)
    if len(ty)==0:
        ty.append(Types[0])

    while ntop:
        if "Base" in results[t].text:
            ntop= False
        t+=1
    for i in range(5):
        sta.append(int(results[t+i].text))
    while len(results[b].text) >= 3 and b<=t:
        if "{}".format(sta[0]) not in results[b].text and "Lv" not in results[b].text:
            mov.append(results[b].text)
        b+=3
    he=float(results[5].contents[2].replace("m", "").strip())
    we = float(results[6].contents[0].replace("lbs", "").strip())
    mons.append(pokemon(nme,img,mfil,ty,mov,cl,biol,cr,sta,he,we,a))
    a+=1
```

# Bios for all pokemons

```python
def bio(pk):
    print(pk.name+" the "+pk.classification+'. '+ pk.bio)
```

# Menus

```python
def menu():
    print("1. Choose pokemon by number")
    print("2. Choose pokemon by name")
    print("3. Choose pokemon by type")
    print("4. Choose pokemon by weight")
    print("5. Choose pokemon by height")
    print("6. Choose pokemon by stats")
    print("7. Choose pokemon by moves")
    print("8. Quit")
```

```python
def main():
    menu()
    c = int(input("Please enter the number of your choice. "))
    while c <1 or c>8:
            c = int(input("Please enter the number of your choice. "))
    while c != 8:    
        if c == 1:
            cho= int(input("Please choose any pokemon (1-151): "))
            while cho <1 or cho >151:
                cho= int(input("Please choose any pokemon (1-151): "))
        elif c == 2:
            in_pk= True
            while in_pk:

                ch= "{}".format(input("Please choose any pokemon (By name): "))
                for i in range (len(mons)):
                    if mons[i].name== ch or mons[i].name.lower()== ch :
                        cho = i+1
                        in_pk = False
                    if i == 150 and in_pk == True:
                        print(ch + " is not registered in the pokedex.")

        elif c == 3:
            pn=[]
            print("{}".format(Types))
            c= "{}".format(input("Please choose any type (The first letter must be captial):  "))
            while c not in Types:
                print("That is not a type.")
                print("{}".format(Types))
                c= "{}".format(input("Please choose any type:  "))
            for i in range (len(mons)):
                if c in mons[i].type:
                    print("{}".format(mons[i].num)+". "+"{}".format(mons[i].name))
                    pn.append(mons[i].num)
            cho= int(input("Please choose any pokemon (By number): "))
            while cho not in pn:
                print('This pokemon does not have '+ c +' typing.')
                cho= int(input("Please choose any pokemon. (By number) "))
        elif c == 4:
            pn=[]
            while len(pn)==0:
                print("Weight should be in between 0.2 lbs to 1014.1 lbs.")
                c= float(input("Please type in the weight of this pokemon (in lbs):  "))
                for i in range (len(mons)):
                    if mons[i].weight==c:
                        print("{}".format(mons[i].num)+". "+"{}".format(mons[i].name))
                        pn.append(mons[i].num)
                if len(pn) == 0:
                    print("None of these Pokemons has weight of "+"{}".format(c)+".")
            cho= int(input("Please choose any pokemon (By number): "))
            while cho not in pn:
                print('This pokemon does not have '+ "{}".format(c) +' lbs.')
                cho= int(input("Please choose any pokemon (By number): "))
        elif c == 5:
            pn=[]
            while len(pn)==0:
                print("Height should be in between 0.2m to 8.8m.")
                c= float(input("Please type in the height of this pokemon (in m):  "))
                for i in range (len(mons)):
                    if mons[i].height==c:
                        print("{}".format(mons[i].num)+". "+"{}".format(mons[i].name))
                        pn.append(mons[i].num)
                if len(pn) == 0:
                    print("None of these Pokemons has height of "+"{}".format(c)+".")
            cho= int(input("Please choose any pokemon (By number): "))
            while cho not in pn:
                print('This pokemon does not have '+"{}".format(c) +' m.')
                cho= int(input("Please choose any pokemon (By number): "))
        elif c == 6:
            pn=[]
            print(st)
            print("We are going to pick a pokemon based on a value of a specifc stat.")
            s= "{}".format(input("Please choose which stat you want to look at. (The first letter must be captial): "))
            while s not in st:
                print("That is not a stat.")
                print(st)
                s= "{}".format(input("Please choose which stat you want to look at: "))
            while len(pn)==0:
                print("Now enter the value of that stat (5-180): ")
                c= int(input(s+": "))
                for i in range (len(mons)):
                    if mons[i].stats[st.index(s)]==c:
                        print("{}".format(mons[i].num)+". "+"{}".format(mons[i].name))
                        pn.append(mons[i].num)
                if len(pn) == 0:
                    print("None of these Pokemons has "+s+" value of "+"{}".format(c)+".")
            cho= int(input("Please choice any pokemon (By number): "))
            while cho not in pn:
                print('This pokemon is not in the list shown.')
                cho= int(input("Please choose any pokemon (By number): "))

        elif c ==7:
            pn=[]
            print("We are going to pick a pokemon that can learn a move you choose.")
            while len(pn) ==0:
                c= "{}".format(input("Please type in the move you want to look at (The first letter must be captial): "))
                for i in range (len(mons)):
                    if c in mons[i].moves:
                        print("{}".format(mons[i].num)+". "+"{}".format(mons[i].name))
                        pn.append(mons[i].num)
                if len(pn) == 0:
                    print("None of these Pokemons can learn "+ c+".")
            cho= int(input("Please choice any pokemon (By number): "))
            while cho not in pn:
                print('This pokemon is not in the list shown.')
                cho= int(input("Please choose any pokemon (By number): "))



        print(mons[cho-1].name)
        with open(mons[cho-1].img,'rb') as file:
            display(Image(file.read()))
        playsound(mons[cho-1].cry)
        print("Type: " + '{}'.format(mons[cho-1].type))
        print("Stats: ")
        for i in range (len(st)):
            print(st[i]+" " +'{}'.format(mons[cho-1].stats[i]))
        print("Catch Rate: "+ '{}'.format(mons[cho-1].catch))
        print("Height: "+ '{}'.format(mons[cho-1].height)+" m")
        print("Weight: "+ '{}'.format(mons[cho-1].weight)+ " lbs")
        bio(mons[cho-1])
        print("Moves they can learn: "+'{}'.format(mons[cho-1].moves))
        menu()
        c = int(input("Please enter the number of your choice. "))
        while c <1 and c>7:
            c = int(input("Please enter the number of your choice. "))
if __name__=='__main__':
    main()
print("Enjoy!!!!")
```
